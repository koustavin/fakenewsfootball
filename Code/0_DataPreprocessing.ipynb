# Further Data Cleaning and Feature Extraction

### Tokenization
The tweets we have right now are basically fully formed sentence(s), at least most of them are. Machine Learning algorithms can't deal with sentences (or even words for that matter) - they mostly deal with numbers. As such, I have to somehow translate our sentences to numbers (or a combination of numbers) - and that's easier to do with words or tokens (group of characters representing a useful semantic unit), rather than sentences. Before that I want to rid our sentences of any punctuation or stopwords (a set of commonly used words in any language, like - the, is, and etc. are stopwords in English) - so that we can concentrate solely on the more 'meaningful' words. The following function has been defined to convert these tweets into a list of tokens, and these are the steps it uses to do that -
1) removing punctuation : a punctuation being a character, I'll have to look at individual characters of the text string (tweet) to compare it with punctuation symbols (from string package), and then join it back to form a string again.
2) tokenize : splitting the string to separate individual words, which would be needed for the next cleaning steps. I will use regex package (re) and pattern '\W+' - which treats all non-words (space, -, \ etc.) as separator. output would be list of tokens (words).
3) remove stopwords : in a logic similar to what was used in the previous step, I'll check each word to see if it's a stopword - I'll get the list of english stopwords from nltk. Probably also worth removing tokens which don't contain any character.
4) clubbing similar words : This is done so that similar words (like - read/reading) are grouped together and not considered as separate entities. There are 2 ways this is usually done : (a) Stemming - removing suffixes, (b) Lemmatizing - grouping together words identified by the word's lemma (canonical form of a set of words). Stemming is faster, but bit mechanical - for instance, meanness/meaning/mean will all be stemmed to mean. Lemmatizing is a bit smarter, since it considers context and always returns a word that's in the dictionary, but lemmatizing is also slower (computationally expensive) and generally returns more number of words than stemming (which may be correct, given the context). Since our dataset isn't very big, I'm going to use lemmatizing.

```python
# import nltk
# nltk.download('stopwords')
# nltk.download('wordnet')
```

```python
stopwords_en = stopwords.words('english')
def clean_n_lemmatize(text):
    '''
        cleans and lemmatizes raw text
        input : raw text
        output : cleaned & lemmatized tokens
    '''
    text_nopunct = ''.join([character for character in text.lower() if character not in string.punctuation])
    tokens = re.split('\W+', text_nopunct)
    tokens_no_stop = [word for word in tokens if ((word not in stopwords_en) & (len(word) > 0))]
    lemmatized_tokens = [wordnetlem().lemmatize(word) for word in tokens_no_stop]
    return lemmatized_tokens
```

```python
df_tweets['tokens'] = df_tweets.tweet.apply(clean_n_lemmatize)
df_tweets.head()
```

### New Features
Other than the tokens themselves, I will try to derive various other numerical features from these tweets that can be helpful in differentiating between real and fake tweets. Note that I am just trying these out - I'll calculate them and check which of these show clearly distinguished patterns for real and fake tweets. Features that I want to try out -
1) length (without punctuation),
2) token% (length of lemmatized and tokenized words - as a percentage of total length from previous feature),
3) stop% (length of stop words - as a percentage of total length from feature 1),
4) punctuation% (length/count of punctuation - as a percentage of total length, i.e., sum of length from feature 1 and length of punctuation, since length from feature 1 above does not contain the length of the punctuation),
5) uppercase% (length/count of uppercase characters - as a percentage of total length from feature 1)

```python
def xtrct_len(text):
    '''
        extract length of text after removing punctuation (before any other cleaning)
        input : unclean text
        output : length of the text (minus punctuation)
    '''
    text_nopunct = ''.join([character for character in text.lower() if character not in string.punctuation])
    return len(text_nopunct)

def xtrct_tokenized_len(token_list):
    '''
        extract sum of length of all the cleaned and lemmatized tokens for each tweet
        input : cleaned tokens
        output : total length (sum) of all the tokens
    '''
    tokenized_length = sum([len(word) for word in token_list])
    return tokenized_length

def xtrct_stop_len(text):
    '''
        extract sum of length of stopwords in text after removing punctuation and tokenizing
        input : unclean text
        output : total length of stopwords in the text
    '''
    text_nopunct = ''.join([character for character in text.lower() if character not in string.punctuation])
    tokens = re.split('\W+', text_nopunct)
    stop_tokens = [word for word in tokens if word in stopwords_en]
    stop_tokens_length = sum([len(stpwrd) for stpwrd in stop_tokens])
    return stop_tokens_length

def xtrct_punct_len(text):
    '''
        extract number of punctuation characters in unclean tweet
        input : unclean tweet
        output : total length of all punctuation in the tweet
    '''
    puncts = [punct for punct in text if punct in string.punctuation]
    return len(puncts)

def xtrct_upper_char_cnt(text):
    '''
        extract number of upper case charatcters in unclean text
        input : unclean text
        output : number of uppercase characters
    '''
    upper_list = re.findall('[A-Z]', text)
    return len(upper_list)
```

```python
df_tweets['len'] = df_tweets.tweet.apply(xtrct_len)
df_tweets['tok_len'] = df_tweets.tokens.apply(xtrct_tokenized_len)
df_tweets['stp_len'] = df_tweets.tweet.apply(xtrct_stop_len)
df_tweets['punct_cnt'] = df_tweets.tweet.apply(xtrct_punct_len)
df_tweets['upr_cnt'] = df_tweets.tweet.apply(xtrct_upper_char_cnt)
df_tweets.head()
```

### Tokenization and Feature Extraction (Take 2)
First I need to change the tokenization function, rather than just removing the punctuation - I will replace them with space, which will help with splitting the string (splitting will take care of consecutive spaces, if any, as well).

The functions specific to feature extraction don't need to be redefined, since they don't contribute to token creation. Only tok_len, tok_pct features need to be recreated - since only those were applied on the tokens (others were applied on the tweets directly).

Also, thinking about it now, I want one more feature - count of tokens in the tokenized and lemmatized list. Let me first test this on a bad tweet.

```python
# Test
text = ' _ crown prince _ al-ahly _ destroys 1\nha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha -ha'
text_nopunct = ''.join([character if character not in string.punctuation else ' ' for character in text.lower()])
text_nopunct_lessha = re.sub('(?:ha *){2,}', 'ha ', text_nopunct)
print(text_nopunct_lessha)
```

```python
def clean_n_lemmatize(text):
    '''
        cleans and lemmatizes raw text
        input : raw text
        output : cleaned & lemmatized tokens
    '''
    # 2 changes :
    # 1) instead of discounting all punctuation (as I was doing earlier), I will now replace punctuation with space
    # text_nopunct = ''.join([character for character in text.lower() if character not in string.punctuation])
    text_nopunct = ''.join([character if character not in string.punctuation else ' ' for character in text.lower()])
    # 2) that many repeated ha-s is not helpful, I'm going to replace 2 or more occurrences of
    # 'ha' or 'ha ' (or more spaces) with a single 'ha '
    text_nopunct_lessha = re.sub('(?:ha *){2,}', 'ha ', text_nopunct)
    tokens = re.split('\W+', text_nopunct_lessha)
    tokens_no_stop = [word for word in tokens if ((word not in stopwords_en) & (len(word) > 0))]
    lemmatized_tokens = [wordnetlem().lemmatize(word) for word in tokens_no_stop]
    return lemmatized_tokens

df_tweets['tokens'] = df_tweets.tweet.apply(clean_n_lemmatize)
df_tweets.head()
```

#### Word Vectorization

I'll use Term Frequency - Inverse Document Frequency (TF-IDF) as my vectorizer. I had done a comparison among CountVectorizer, TF-IDF, and N-grams with Naive Bayes here : https://www.kaggle.com/code/koustavin/94-acc-multinomialnb-with-diff-vectorizers - the results were comparable. This time, I'm going with the combination of TF-IDF and N-grams (a combination of Unigrams, Bigrams, and Trigrams) with a maximum feature cap of 35000.

I wanted to limit my feature count - since I have roughly 36,000 data samples in our training set, I chose 35,000 features in TF-IDF. I expect to lose a little bit information with this limit, but bringing down my feature count to less than sample count should improve generalization. I am hoping that using a combination of TF-IDF and N-grams would balance the lost information. I have also tried word embedding (Word2Vec) with positional embedding (notebook should be available in Word2Vec path), but the results with Logistic Regression without any hyperparameter tuning was marginally worse than what we currently have.

That's not to say that this is the best combination possible, specifically when certain word vector may perform better for certain classification methods, but using multiple vectorization methods in parallel with multiple classification algorithms will make comparing the final scores very complicated.

I am still going to use the tokens that I had generated, but N-grams requires concatenated string -  so, I will join those tokens in a sentence like string, separated by space.

```python
X_train['rejoined_tokens'] = X_train.tokens.apply(lambda tokens: ' '.join(tokens))
X_test['rejoined_tokens'] = X_test.tokens.apply(lambda tokens: ' '.join(tokens))
```

##### TF-IDF

```python
tfidf_unibitrigram = TfidfVectorizer(ngram_range = (1, 3), max_features = 35000)
vectorized_sparse_matrix_train = tfidf_unibitrigram.fit_transform(X_train['rejoined_tokens'])
vectorized_sparse_matrix_test = tfidf_unibitrigram.transform(X_test['rejoined_tokens'])
```

#### Merging Word Vectors and Scaled Numerical Features

Now that I have our vectorized tweet data in sparse matrix format, I need to combine it with our scaled numerical features - and that will be our final dataset, to be fed to the ML models.

At first, I was trying to convert this sparse matrix to a dataframe and then to concatenate it with my numerical features. But that spiked my vm's (running Jupyter) memory consumption from usual 2 GB to around 10 GB and notebook kernel kept crashing. Finally decided to go the other way - converted my numrical features to sparse matrix format, and then stacked them horizontally. Not only did that take care of the memory issue, this actually improved ML algo speed as well - noticed significant drops in training and prediction times.

```python
num_feat_scaled_sparse_train = sparse.csr_matrix(X_train[numerical_features_scaled].values)
num_feat_scaled_sparse_test = sparse.csr_matrix(X_test[numerical_features_scaled].values)

sparse_train = sparse.hstack((num_feat_scaled_sparse_train, vectorized_sparse_matrix_train))
sparse_test = sparse.hstack((num_feat_scaled_sparse_test, vectorized_sparse_matrix_test))
```
